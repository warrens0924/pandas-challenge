```python
# Dependencies and Setup
import pandas as pd
import os
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = os.path.join('Resources/schools_complete.csv')
student_data_to_load = os.path.join('Resources/students_complete.csv')

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left",on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#id unique values of schools

total_schools = len(school_data_complete['school_name'].unique())
#total students
students = school_data_complete['student_name'].count() 
#total budget
budget_total = school_data['budget'].sum()

#average score
math_avg = school_data_complete['math_score'].mean()

#average score
reading_avg = school_data_complete['reading_score'].mean()
#passing & passing percent                                        
passing_math = len(school_data_complete.loc[school_data_complete['math_score'] >=70])
m_pass_pct = passing_math/students  * 100
#passing & passing percent                                      
passing_reading = len(school_data_complete.loc[school_data_complete['reading_score'] >=70])
r_pass_pct = passing_reading/students * 100
#combined percentage
overall_pct = student_data[(student_data['math_score'] >=70) & (student_data['reading_score'] >= 70)]['student_name'].count()/students * 100 





```

```python

#dataframe
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
"Total Students": [students], "Total Budget": [budget_total], "Average Math Score": [math_avg],
            "Average Reading Score": [reading_avg], "Math Pass Percentage": [m_pass_pct],
                    "Reading Pass Percentage": [r_pass_pct],
                        "Combined Percentage": [overall_pct]
})

district_summary_df["Total Budget"]=district_summary_df['Total Budget'].map("${:,.2f}".format)
```
